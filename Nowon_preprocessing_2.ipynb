```python
dt = pd.read_csv('C:/Users/1102j/anaconda/dataScience/project/nowon_apt.csv', encoding='euckr')
dt
```

```python
dt['일반관리-인원']
```

```python
data = pd.DataFrame()
data
```

```python
data['단지명(K-apt)'] = dt['단지명']
```

```python
data['일반관리'] = dt['일반관리-인원']
data['경비관리'] = dt['경비관리-인원']
data['청소관리'] = dt['청소관리-인원']
```

```python
data
```

```python
m_merge = pd.merge(m, data, how='left', on='단지명(K-apt)')
len(m_merge)
```

```python
y_merge = pd.merge(y, data, how='left', on='단지명(K-apt)')
len(y_merge)
```

```python
m_merge.to_csv('month_nowon.csv', encoding='euckr', index=False)
```

```python
y_merge.to_csv('year_nowon.csv', encoding='euckr', index=False)
```
